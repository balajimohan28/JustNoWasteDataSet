```python
emissions_vic_df = pd.read_excel('emissionsVic.xlsx', sheet_name='Sheet1')
```

```python
emissions_land_facility_id_df = emissions_vic_df.groupby('facility_id')["land_emission_kg"].sum().reset_index(name ='Total_Land_Waste_Amount')
```

```python
emissions_air_point_facility_id_df = emissions_vic_df.groupby('facility_id')["air_point_emission_kg"].sum().reset_index(name ='Total_Air_Point_Waste_Amount')
emissions_air_fugitive_facility_id_df = emissions_vic_df.groupby('facility_id')["air_fugitive_emission_kg"].sum().reset_index(name ='Total_Air_Fugitive_Waste_Amount')
emissions_air_total_facility_id_df = emissions_vic_df.groupby('facility_id')["air_total_emission_kg"].sum().reset_index(name ='Total_Air_Waste_Amount')
emissions_water_facility_id_df = emissions_vic_df.groupby('facility_id')["water_emission_kg"].sum().reset_index(name ='Total_Water_Waste_Amount')
```

```python
emissions_land_facility_id_df.sort_values('facility_id').reset_index()
emissions_land_facility_id_df.head()
```

```python
emissions_air_point_facility_id_df.sort_values('facility_id').reset_index()
emissions_air_fugitive_facility_id_df.sort_values('facility_id').reset_index()
emissions_air_total_facility_id_df.sort_values('facility_id').reset_index()
emissions_water_facility_id_df.sort_values('facility_id').reset_index()
emissions_land_facility_id_df.head()
```

```python
merged_left = pd.merge(left=facility_df,right=emissions_land_facility_id_df, how='left', left_on='facility_id', right_on='facility_id')
```

```python
merged_left1 = pd.merge(left=merged_left,right=emissions_air_point_facility_id_df, how='left', left_on='facility_id', right_on='facility_id')
merged_left2 = pd.merge(left=merged_left1,right=emissions_air_fugitive_facility_id_df, how='left', left_on='facility_id', right_on='facility_id')
merged_left3 = pd.merge(left=merged_left2,right=emissions_air_total_facility_id_df, how='left', left_on='facility_id', right_on='facility_id')
merged_left4 = pd.merge(left=merged_left3,right=emissions_water_facility_id_df, how='left', left_on='facility_id', right_on='facility_id')
```

```python
merged_left4.to_csv("facility_with_emissions_kg.csv", index = False)
```
